```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap


import warnings 
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler, normalize

from sklearn.cluster import KMeans

import scipy.cluster.hierarchy as sch
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering


from sklearn.decomposition import PCA

from sklearn.metrics import silhouette_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.rename(columns={"SepalLengthCm":"sep_l_cm","SepalWidthCm":'sep_w_cm',
                   'PetalLengthCm':'pet_l_cm','PetalWidthCm':'pet_w_cm',
                   'Species':'spec'},inplace=True)
```

```python
X=df.iloc[:,1:5]
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
pca = PCA(n_components = 2)
X = pca.fit_transform(X)
explained_variance = pca.explained_variance_ratio_
```

```python
fig = plt.figure(figsize = (10,5))
plt.scatter(X[:,0],X[:,1],s=100,c='magenta',label='All customers')
plt.title('Species before clustering')
plt.xlabel('PCA_0')
plt.ylabel('PCA_1')
plt.legend()
plt.show()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
```

```python
X_km=kmeans.fit_predict(X)
```

```python
#Lets plot those clusters!
fig = plt.figure(figsize = (15,8))
plt.scatter(X[X_km==0,0],X[X_km==0,1],s=100,c='magenta')
plt.scatter(X[X_km==1,0],X[X_km==1,1],s=100,c='blue')
plt.scatter(X[X_km==2,0],X[X_km==2,1],s=100,c='green')
plt.scatter(X[X_km==3,0],X[X_km==3,1],s=100,c='black')
plt.scatter(X[X_km==4,0],X[X_km==4,1],s=100,c='burlywood')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='red',label='Centroids')
plt.title('Species after clustering')
plt.xlabel('Pca_0')
plt.ylabel('Pca_1')
plt.legend()
plt.show()
```

```python
dendogram=sch.dendrogram(sch.linkage(X,method='ward')) # Within cluster variance is reduced with ward method
plt.title('Dendogram')
plt.xlabel('Irises')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
fig = plt.figure(figsize = (15,8))
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='magenta')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='blue')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='green')
plt.title('Cluster of Species')
plt.xlabel('PCA_0')
plt.ylabel('PCA_1')
plt.ioff()
plt.show()
```
